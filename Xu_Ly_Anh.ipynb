# Resize ảnh

```python
import cv2
import os

# Đường dẫn thư mục gốc
input_folder = r"E:/NCKH/Data/fall_dataset/images/train"
output_folder = r"E:/NCKH/Data/fall_dataset/img_train"

# Tạo thư mục lưu ảnh resized nếu chưa có
os.makedirs(output_folder, exist_ok=True)

# Kích thước ảnh muốn resize
target_size = (96, 96)

# Duyệt tất cả ảnh trong thư mục input
for filename in os.listdir(input_folder):
    if filename.lower().endswith((".jpg", ".png", ".jpeg", ".bmp")):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)

        # Đọc ảnh và resize
        img = cv2.imread(input_path, cv2.IMREAD_GRAYSCALE)
        if img is not None:
            img_resized = cv2.resize(img, target_size)
            cv2.imwrite(output_path, img_resized)
            print(f"Đã xử lý: {filename}")
        else:
            print(f"Lỗi đọc ảnh: {filename}")
```

```python
import cv2
import os
import numpy as np
import random

# Thư mục ảnh gốc
input_folder = r"E:/NCKH/Data/fall_dataset/img_train"
output_folder = input_folder  # Hoặc đổi thành thư mục mới nếu muốn tách

os.makedirs(output_folder, exist_ok=True)

# Hàm xoay ảnh
def rotate_image(image, angle):
    h, w = image.shape[:2]
    M = cv2.getRotationMatrix2D((w/2, h/2), angle, 1)
    return cv2.warpAffine(image, M, (w, h), borderMode=cv2.BORDER_REFLECT)

# Hàm thay đổi độ sáng
def change_brightness(image, factor=1.25):
    return cv2.convertScaleAbs(image, alpha=factor, beta=0)

# Hàm thêm nhiễu Gaussian
def add_gaussian_noise(image, mean=0, std=15):
    noise = np.random.normal(mean, std, image.shape).astype(np.uint8)
    noisy = cv2.add(image, noise)
    return noisy

# Duyệt từng ảnh
for filename in os.listdir(input_folder):
    if filename.lower().endswith((".jpg", ".png", ".jpeg", ".bmp")):
        path = os.path.join(input_folder, filename)
        img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            print(f"Lỗi ảnh: {filename}")
            continue

        name, ext = os.path.splitext(filename)

        # 1. Xoay ±15 độ
        rotated = rotate_image(img, random.choice([-15, 15]))
        cv2.imwrite(os.path.join(output_folder, f"{name}_rotated{ext}"), rotated)

        # 2. Làm mờ nhẹ
        blurred = cv2.GaussianBlur(img, (5, 5), 0)
        cv2.imwrite(os.path.join(output_folder, f"{name}_blurred{ext}"), blurred)

        # 3. Tăng độ sáng
        bright = change_brightness(img, factor=1.3)
        cv2.imwrite(os.path.join(output_folder, f"{name}_bright{ext}"), bright)

        # 4. Lật ngang
        flipped = cv2.flip(img, 1)
        cv2.imwrite(os.path.join(output_folder, f"{name}_flipped{ext}"), flipped)

        print(f"Đã tạo ảnh tăng cường cho: {filename}")
```

# Tạo dữ liệu gán nhãn 1 (Ngã)

```python
import csv

folder_path = r"E:/NCKH/Data/fall_dataset/img_train"
csv_path = os.path.join(folder_path, "image_labels.csv")

with open(csv_path, mode='w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["filename", "label"])

    for filename in os.listdir(folder_path):
        if filename.lower().endswith(('.jpg', '.png', '.jpeg', '.bmp')):
            if filename.lower().startswith("fall"):
                label = 1
            elif filename.lower().startswith("not fallen"):
                label = 0
            else:
                continue  # bỏ qua ảnh không rõ nhãn

            writer.writerow([filename, label])

print("✅ Đã tạo file image_labels.csv thành công!")
```
